```python
#empty list 

Review = []

for i in range(1,2):
    ip =[]
    url = "https://www.flipkart.com/sony-ps4-slim-1000-gb-spider-man-ratchet-clank-gran-turismo/p/itmb08a36626ba21?pid=GMCFUAKAMHX6QJ2C&lid=LSTGMCFUAKAMHX6QJ2C1HTDLX&marketplace=FLIPKART&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&fm=SEARCH&iid=8ce00642-7da8-42e0-ac17-85ddfea304ab.GMCFUAKAMHX6QJ2C.SEARCH&ppt=sp&ppn=sp&ssid=qoigwdv4og0000001609702840467&qH=28be67467aca7ee6"+str(i)
    response = requests.get(url)
    soup = bs(response.content,"html.parser")
    review = soup.findAll("span",{"class":""})
    for i in range(len(review)):
        ip.append(review[i].text)
    Review = Review+ip
```

```python
# combined review 
import re
Review_combined = " ".join(Review)


Review_combined = re.sub("[^A-Za-z" "]+"," ",Review_combined).lower()
Review_combined = re.sub("[0-9" "]+"," ",Review_combined)


Review_words = Review_combined.split(" ")
stop_words = stopwords.words("english")
Review_words =[w for w in Review_words if not w in stop_words]
```

```python
import matplotlib.pylab as plt
from wordcloud import WordCloud

with open("C:\\Data science\\Data sets\\TEXT MINING\\stop.txt","r") as sw:
    stop_word = sw.read()
    
  
    Review_word = [w for w in Review_words if not w in stop_word]
    
    Review_stopwords = " ".join(Review_word)
```

```python
wordcloud_st = WordCloud(
                      background_color='black',
                      width=1800,
                      height=1400
                     ).generate(Review_stopwords)
plt.imshow(wordcloud_st)
```
