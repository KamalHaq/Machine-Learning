```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# callig simple matplot functions for simple plots
# data preperation
x = np.linspace(0,10,50)
plt.plot(x,x,label='Linear line')
# adding legend
plt.legend()
#Visulaising the plot
plt.show()
```

```python
#Object oriented functions approach
fig,ax = plt.subplots(figsize=(5,5))
ax.bar(x=['X1','X2','X3'],height=[3,7,4],color='b')
ax.set_xlabel(xlabel="X TITLE",color='r',size=22)
ax.set_ylabel(ylabel="Y TITLE",color='c',size=30)
plt.show()
```
