```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
test = pd.read_csv("C:\\Data science\\Data sets\\Naive Bayes\\SalaryData_Test.csv")
test.head()
```

```python
test.drop(["Age","Education"],axis="columns")
test.head()
```

```python
test["dummy"]=pd.factorize(test.Workclass)[0]
test["dummy1"]=pd.factorize(test.Maritalstatus)[0]
test["dummy2"]=pd.factorize(test.Occupation)[0]
test["dummy3"]=pd.factorize(test.Relationship)[0]
test["dummy4"]=pd.factorize(test.Sex)[0]
test["dummy5"]=pd.factorize(test.Native)[0]
test["dummy6"]=pd.factorize(test.Salary)[0]
test["dummy7"]=pd.factorize(test.Race)[0]
test1=test.drop(["Workclass","Maritalstatus","Occupation","Relationship","Race","Sex","Native","Salary"],axis=1)
test2=test1.rename(columns={"dummy":"Workclass","dummy1":"Maritalstatus","dummy2":"Occupation","dummy3":"Relationship",
                    "dummy4":"Sex","dummy5":"Native","dummy6":"Salary","dummy7":"Race"})
```

```python
test2.head()
```

```python


def dnorm(i):
    x = (i-i.min())/i.max()-i.min()
    return(x)
norm_data = dnorm(test2.iloc[:,:])

norm_data
```

```python
target=norm_data.Salary
inputs=norm_data.drop(["Salary"],axis="columns")
```

```python
#no NA values
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
len(x_train)
#len(x_test)
```

```python
len(inputs)
```

```python
#Naive  Bayes model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
